```python
import pandas as pd
import sqlite3
import json
orders=pd.read_csv('./orders.csv')
useres=pd.read_json('./users.json')

connection=sqlite3.connect(':memory:')
with open('./restaurants.sql','r') as file:
    connection.executescript(file.read())

    restaurants=pd.read_sql_query('SELECT * FROM restaurants',connection)

    merged_with_users=pd.merge(orders,useres,on='user_id',how='left')
    final_data=pd.merge(merged_with_users,restaurants,on='restaurant_id',how='left')
    final_data.to_csv('final_data.csv',index=False)

    print(final_data.head())
```
